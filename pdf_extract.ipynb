```python
import pytesseract
from PIL import Image, ImageEnhance
from pdf2image import convert_from_path

tesseract_path = "C:\\Program Files\\Tesseract-OCR\\tesseract.exe"
def extract_enhanced_text(pdf_file):
    images = convert_from_path(pdf_file)
    extracted_text = ""
    for i, image in enumerate(images):
        enhancer_sharpness = ImageEnhance.Sharpness(image)
        image_sharp = enhancer_sharpness.enhance(3.2)
        #enhancer_contrast = ImageEnhance.Contrast(image_sharp)
        #image_sharp_bright = enhancer_contrast.enhance(1.2)
        pytesseract.pytesseract.tesseract_cmd = tesseract_path
        text = pytesseract.image_to_string(image_sharp)
        extracted_text += text
    return (extracted_text)

new_enhanced_text = extract_enhanced_text("1_sample_2.pdf") 
with open("enhanced_text2.txt","w") as file:
    file.write(new_enhanced_text)
```

```python
import re

the_text = ""
with open ("enhanced_text2.txt", "r") as file_hnd:
    for line in file_hnd:
        the_text = the_text + line

blocks = re.split(r'Constituency', the_text)

# The first element of the 'blocks' list will be empty, so remove it
blocks = blocks[1:]
#print(blocks)
# Initialize variables
data = []
serial_no = 1

# Define regular expressions to capture desired key-value pairs
regex_patterns = {
    "Name": r"(?<![ds]\s)Name\s*[:!;?+#1$}%&()\s^@*/|-]\s*(?!(?:Name|Fathers|Husbands|House|Age|Gender)\s*)(\w+(?:\s+\w+){0,2})",
    "Fathers Name": r"Fathers Name\s*[:!;?}+#1$%&()\s^@*/|-]*\s*((?!(?:Name|Fathers|Husbands|House|Age|Gender)\b)\w+(?:\s+\w+){0,2})",
    "Husbands Name": r"Husbands Name\s*[:!;?+1}#$%&()\s^@*/|-]*\s*((?!(?:Name|Fathers|Husbands|House|Age|Gender)\b)\w+(?:\s+\w+){0,2})",
    "House Number": r"House Number\s*[:!;?}+#$%1&()\s^@*/|-]*\s*([\w/-]+)",
    "Age": r"Age\s*[:!;?+#$%}&1()\s^@*\/|-]*\s*(\d+)",
    "Gender": r"Gender\s*[:!;?+1}#$%&()\s^@*\/|-]*\s*(\w+)",
}
id_pattern = r'\b(?:KGY|KGY\s)?\d{6,10}\b|\b(?:TD[0ZzDdOo2]|To[0zZDdOo2]|T0[0ZzDdOo2]|TO[0ZzDdOo2]\s)?\d{6,10}\b'

#for block in blocks:
    #print(block.split("\n"))
# Process each block and extract desired key-value pairs
for block in blocks:
    person_data = {"Serial No": [], "ID": [], "Name":[], "Fathers Name": [], "Husbands Name": [], 
                   "Gender": [], "Age": [], "House Number": [] }

    for line in block.split("\n"):   
        father_name_found = False
        husband_name_found = False  
        for key, pattern in regex_patterns.items():
            match = re.search(pattern, line)
            #print(match)
            if match:
                if match.group(0).split(":")[0].strip() == "Fathers Name":
                    father_name_found = True
                    #husband_name_found= False
                elif match.group(0).split(":")[0].strip() == "Husbands Name":
                    husband_name_found= True
                    #father_name_found = False
                if father_name_found == True:
                    person_data["Fathers Name"].append(match.group(1))
                    person_data["Husbands Name"].append(None)
                elif husband_name_found == True:
                    person_data["Husbands Name"].append(match.group(1))
                    person_data["Fathers Name"].append(None)
                else:
                    person_data[key].append(match.group(1))

    for lines in block.split("\n"):
        id_match = re.search(id_pattern, lines)
        if id_match:
            person_data["ID"].append(id_match.group(0))

    #if person_data:
        #person_data["Serial No"].append(serial_no)
        #serial_no += 1
    data.append(person_data)
        
#max_items = max(len(data[0].get("ID", [])) for data_dict in data)
#serial_numbers = list(range(1, max_items + 1))
for data_dict in data:
    for item in data_dict["Name"]:
        data_dict["Serial No"].append(serial_no)
        serial_no += 1
# Print the cleaned data
for person in data:
    print(person)
```

```python
def id_count(text_file):
    the_text = []
    ids = []
    with open (text_file) as file_hnd:
        for line in file_hnd:
            line = line.rstrip()
            the_text.append(line)
        count = 0
        for line in the_text:
            words = line.split()
            for word in words:
                if (word.startswith("TD")) or (word.startswith("KG")):
                    ids.append(word)
                    count +=1
    return (count,ids)
print(id_count("enhanced_text.txt"))
```
